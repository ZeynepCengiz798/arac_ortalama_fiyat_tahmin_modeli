TensorFlow Keras ile embedding‚Äôli modelimizi kuralƒ±m

```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, Concatenate
from tensorflow.keras.models import Model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# üîπ Sayƒ±sal ve kategorik s√ºtunlarƒ± ayƒ±r
numeric_cols = ['km', 'myear','marka_model_yil_avg', 'relative_price']
categorical_cols = [c + '_enc' for c in cat_cols]
```

```python
# üîπ Sayƒ±sal veriyi standardize et
scaler = StandardScaler()
X_numeric_scaled = scaler.fit_transform(df[numeric_cols])
```

```python
# üîπ Kategorik veriyi al
X_categoricals = [df[col].values for col in categorical_cols]
```

```python
# üîπ Hedef deƒüi≈üken
y = df['listed_price'].values
```

```python
# üîπ Train-test b√∂l
X_numeric_train, X_numeric_test, y_train, y_test = train_test_split(
    X_numeric_scaled, y, test_size=0.2, random_state=42)

X_categoricals_train = [col[:len(y_train)] for col in X_categoricals]
X_categoricals_test = [col[len(y_train):] for col in X_categoricals]
```

```python
# üîπ Model giri≈üleri
inputs = []
encoded_layers = []
```

```python
# Sayƒ±sal giri≈ü
input_numeric = Input(shape=(len(numeric_cols),), name="numeric_input")
inputs.append(input_numeric)
encoded_layers.append(input_numeric)
```

```python
# Kategorik giri≈üler + embedding katmanlarƒ±
for i, col in enumerate(categorical_cols):
    input_cat = Input(shape=(1,), name=col)
    vocab_size = df[col].nunique() + 1
    embed_dim = min(50, vocab_size // 2)
    embedding = Embedding(input_dim=vocab_size, output_dim=embed_dim, name=f"embed_{col}")(input_cat)
    flat = Flatten()(embedding)
    inputs.append(input_cat)
    encoded_layers.append(flat)
```

```python
# üîπ T√ºm giri≈üleri birle≈ütir
x = Concatenate()(encoded_layers)
x = Dense(64, activation='relu')(x)
x = Dense(32, activation='relu')(x)
output = Dense(1)(x)
```

```python
# üîπ Modeli tanƒ±mla

model = Model(inputs=inputs, outputs=output)
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
model.summary()
```

```python
# üîπ Modeli eƒüit
model.fit(
    [X_numeric_train] + X_categoricals_train,
    y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.1
)
```

```python
loss, mae = model.evaluate([X_numeric_test] + X_categoricals_test, y_test)
print(f"Test Mean Absolute Error (MAE): {mae:.2f}")
```

```python
y_pred = model.predict([X_numeric_test] + X_categoricals_test)

# ƒ∞lk 5 tahmin ile ger√ßek fiyatƒ± kar≈üƒ±la≈ütƒ±ralƒ±m
import numpy as np
for i in range(5):
    print(f"Ger√ßek Fiyat: {y_test[i]:,.0f} TL - Tahmin: {y_pred[i][0]:,.0f} TL")
```

```python
import numpy as np

# Hedef deƒüi≈ükeni log1p ile d√∂n√º≈üt√ºr (log(1 + listed_price))
y_log = np.log1p(df['listed_price'].values)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

numeric_cols = ['km', 'myear','marka_model_yil_avg', 'relative_price']
categorical_cols = [c + '_enc' for c in cat_cols]

X_numeric_scaled = StandardScaler().fit_transform(df[numeric_cols])
X_categoricals = [df[col].values for col in categorical_cols]
```

```python
X_numeric_train, X_numeric_test, y_train, y_test = train_test_split(
    X_numeric_scaled, y_log, test_size=0.2, random_state=42)

X_categoricals_train = [col[:len(y_train)] for col in X_categoricals]
X_categoricals_test = [col[len(y_train):] for col in X_categoricals]
```

```python
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, Concatenate
from tensorflow.keras.models import Model
```

```python
inputs = []
encoded_layers = []
```

```python
# Sayƒ±sal giri≈ü
input_numeric = Input(shape=(len(numeric_cols),), name="numeric_input")
inputs.append(input_numeric)
encoded_layers.append(input_numeric)
```

```python
# Kategorik giri≈ü + embedding
for col in categorical_cols:
    input_cat = Input(shape=(1,), name=col)
    vocab_size = df[col].nunique() + 1
    embed_dim = min(50, vocab_size // 2)
    embedding = Embedding(input_dim=vocab_size, output_dim=embed_dim)(input_cat)
    flat = Flatten()(embedding)
    inputs.append(input_cat)
    encoded_layers.append(flat)
```

```python
# Katmanlarƒ± birle≈ütir
x = Concatenate()(encoded_layers)
x = Dense(64, activation='relu')(x)
x = Dense(32, activation='relu')(x)
output = Dense(1)(x)

model = Model(inputs=inputs, outputs=output)
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
# Modeli eƒüit
model.fit(
    [X_numeric_train] + X_categoricals_train,
    y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.1
)
```

```python
# Deƒüerlendirme
loss, mae = model.evaluate([X_numeric_test] + X_categoricals_test, y_test)
print(f"Loglu MAE: {mae:.2f}")
```

```python
# Tahmin
y_pred_log = model.predict([X_numeric_test] + X_categoricals_test)
y_pred = np.expm1(y_pred_log)  # log1p'nin tersini alƒ±yoruz
y_true = np.expm1(y_test)
```

```python
def plot_actual_vs_predicted(y_true, y_pred):
    plt.figure(figsize=(8, 8))
    plt.scatter(y_true, y_pred, alpha=0.5, color='blue')
    plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--', lw=3)  # 45 derece referans √ßizgisi
    plt.xlabel("Ger√ßek Fiyatlar (‚Ç∫)")
    plt.ylabel("Tahmin Edilen Fiyatlar (‚Ç∫)")
    plt.title("Ger√ßek vs Tahmin Edilen Ara√ß Fiyatlarƒ±")
    plt.grid(True)
    plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

def print_model_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    print(f"MAE (Ortalama Mutlak Hata): {mae:,.0f} ‚Ç∫")
    print(f"RMSE (Karek√∂k Ortalama Kare Hata): {rmse:,.0f} ‚Ç∫")
```
